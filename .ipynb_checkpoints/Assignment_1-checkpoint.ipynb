# 1) Area of a Rectangle

```python
length = float(input("Enter a Length"))
width =  float(input("Enter a Width"))

Area = length * width
print("The Area Of Rectangle is: ", round(Area,2))
```

# 2) Circumference of a Circle

```python
import math

r = float(input("Enter a Radius: "))
Circumference = 2* math.pi * r
print("The Circumference of Circle: ", round(Circumference,2))
```

# 3) Simple Interest

```python
Principal = float(input("Enter Principal: "))
Rate = float(input("Enter Rate: "))
Time = int(input("Enter Time: "))

simple_interest = Principal * Rate * Time

print("Simpe Interset: ", round(simple_interest,2))
```

# 4) Speed of an Object

```python
Distance = float(input("Enter Distance: "))
Time = int(input("Enter Time: "))

Speed = Distance / Time
print("Speed of an Object: ", Speed)
```

# 5) BMI Calculator

```python
Weight = float(input("Enter Weight in (Kg): "))
Height = float(input("Enter Height in (Meters): "))

BMI = Weight / (Height**2)
print("Body Mass Index: ", round(BMI,2))
```

# 6) Force Using Newton Law

```python
m = float(input("Enter mass in (Kg): "))
meter = float(input("Enter the value of meter: "))
time_in_sec = float(input("Enter the time in seconds: "))

a = meter / time_in_sec ** 2

F = m * a

print("The Force of an Object : ", F)
```

# 7) Compound Interest

```python

P = float(input("Enter the principal amount: "))
r = float(input("Enter the annual interest rate (decimal): "))
n = float(input("Enter the number of times interest is compounded per year: "))
t = float(input("Enter the time in years: "))
A = P * (1 + r/n) ** (n-t)
print("Compounc Interest: ", round(A,2))
```

# 9) Volume of a Sphere

 # 10) Kinetic Energy

```python
m = float(input("Enter mass in (Kg): "))
meter = float(input("Enter the value of meter: "))
second = float(input("Enter the time in seconds: "))

V = meter/second

KE = 1/2 * m * (V ** 2)

print("Kinetic energy of an object: ", round(KE,2))
```

# 11) Quadratic Equation Roots

```python
import cmath

a = int(input("Enter a : "))
b = int(input("Enter b : "))
c = int(input("Enter c : "))

X1 = -b + cmath.sqrt(b**2 - 4*a*c) / 2*a
X2 = -b - cmath.sqrt(b**2 - 4*a*c) / 2*a

print("The roots are", X1 , "and" , X2)
```

# 13) Gravitational Force

```python
G = 6.674e-11 
m1 = float(input("Enter mass of Object 1 : "))
m2 = float(input("Enter mass of Object 2 : "))
r = float(input("Enter distance between the objects : "))

F = G*((m1 * m2)/r**2)

print("The Gravitational Force B/w two Objects: ", F)
```

# 18) Future Value in Saving

```python
PV = float(input("Enter a Present Value : "))
r = float(input("Enter a Radius : "))
t = int(input("Enter a time (years) : "))

FV = PV * ((1 + r)**t)

print("The Future Value of an Investment", round(FV, 2))
```

# 19) Work Done by a Force

```python
import math

f =  float(input("Enter a Force: "))
d =  float(input("Enter a distance: "))
theta = float(input("Enter a theta: "))

W = f * d * math.cos(theta)
print("Work Done: ", round(W,2))
```

# 20) Heat Transfer

```python
m =  float(input("Enter a mass: "))
c =  float(input("Enter a specific heat capacity: "))
t1 = float(input("Enter a t1 (heat): "))
t2 = float(input("Enter a t2 (heat): "))

delta_T = t2 - t1

Q = m * c * delta_T

print("Amount of Heat Transferred: ", round(Q,2))
```
